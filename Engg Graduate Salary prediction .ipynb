# cleaning specialization column

```python
df['Specialization'].unique()
```

```python
df['Specialization'].value_counts()
```

```python
specialization=df.Specialization.value_counts()
```

```python
spe_less_than_10=specialization[specialization<=10]
spe_less_than_10
```

```python
def handle_specialization(value):
    if value in spe_less_than_10:
        return 'other'
    else:
        return value
    
df['Specialization']=df.Specialization.apply(handle_specialization)
```

# ML MOdel

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender']=le.fit_transform(df.Gender)
df['Degree']=le.fit_transform(df.Degree)
df['Specialization']=le.fit_transform(df.Specialization)
df.tail()
```

```python
x=df[['10percentage','12percentage','CollegeTier','Degree','Specialization','collegeGPA']]

y=df['Salary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x)
```

```python
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LinearRegression()
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
```

```python
predictions[:5]
```

```python
y_test[:5]
```

```python
model.score(x_test,y_test)
```
